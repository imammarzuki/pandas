```python
import pandas as pd
```

```python
grades1=pd.read_csv("grades1.csv")
grades2=pd.read_csv("grades2.csv")
students=pd.read_csv("students.csv")
contact=pd.read_csv("contact.csv")
```

```python
grades1["Semester"]=1
```

```python
grades1
```

```python
grades2["Semester"]=2
```

```python
grades2
```

```python
#merger secara vertikal
```

```python
grades=pd.concat([grades1,grades2])
```

```python
grades
```

```python
grades=grades.reset_index(drop=True)
```

```python
grades
```

```python
#merger horizontal
students
```

```python
students_grades=pd.merge(students,grades,left_on="id",right_on="student_id")
```

```python
students_grades
```

```python
students2=students.rename({"id":"student_id"}, axis=1)
```

```python
students2
```

```python
students2_grades=students2.merge(grades,on="student_id")
```

```python
students2_grades
```

```python
students3 = students.set_index("id")
contacts3 = contact.set_index("student_id")

students3_contact3=students3.join(contacts3)
```
